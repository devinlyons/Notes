# Linear Algebra

## Vectors
Vectors represent movement along axes in the order of x, y, z. Consider the vector:

$\left[
\begin{matrix}
1 \\
5 \\
-2 \\
\end{matrix}
\right]$

This vector represents moving 1 to the right on the x-axis, 5 up on the y-axis and 2 back on the z-axis.

Each element in a vector is called an *element* or *coordinate*.
A vector of _n_ real elements defines an _n_ dimensional vector and belongs to $\mathbb{R}^n$.
Such a vector can be said: $\vec{x} \in\mathbb{R}^n$

Vectors can be transposed from typical column-vectors into row-vectors. Transposition is given the symbol *T*.

$\left[
\begin{matrix}
x_1 \\
x_2 \\
... \\
x_n \\
\end{matrix}
\right]^T = [x_1 x_2 ... x_n]$

The magnitude of a vector can be found with the pythagorean theorem.

$||x|| = \sqrt{x_1^2 + x_2^2 + ... + x_n^2}$

The direction of a vector can be found with the _arctan_.

### Vector Addition

$\vec{x} + \vec{y} = 
\left[
\begin{matrix}
x_1 + y_1 \\
x_2 + y_2 \\
... \\
x_n + y_n \\
\end{matrix}
\right]$

### Vector Scalar Multiplication

$a\vec{x} = 
\left[
\begin{matrix}
ax_1 \\
ax_2 \\
... \\
ax_n \\
\end{matrix}
\right]$

## Matrix
A Matrix is a two dimensional array that contains the same elements as the vector. 
A matrix can have $m$ rows and $n$ columns.
The element of matrix $a$ would be identified by $a_{ij}$.
Matrices can be added to each other.
They can also be multiplied by another matrix or by a scaler.

### Matrix Addition
Two matrices with the same dimensions can be added, C = A + B, such that each element of C has the value $c_{ij} = a_{ij} + b_{ij}$.

### Matrix Scaler Multiplication
A matrix can be multiplied by a scaler value such that each element of matrix $B$ in $B = x \times A$ has a value of $b_{ij} = x \times a_{ij}$.

### Matrix Multiplication
Two Matrices can also be multiplied together provided that they are compatible.
Matrices $A$ and $B$ can by multiplied, $C = A \times B$ if $A$ has dimensions $t \times m$ and $B$ has dimensions $m \times v$.
This will produce a matrix of dimensions $t \times v$.
Note that the order the matrices appear in the equation matters.
This can be thought of as $A$ transforms $B$ into $C$.
