# Data Visualization

Qualitative data regards the qualities something has while quantitative data deals with numeric data.
Qualitative data can also be called categorical.
Quantitative data can be discrete or continuous.
Nominal is data with no inherent ordering.
Ordinal has an inherent ordering.
Transforms like `log` may be used in data visualization but the tick marks will need to be adjusted.

## Faceting
Faceting shows multiple graphs side by side.
Typically, each graph will represent a single category.

Faceting is good for breaking down complicated data into simpler parts.
One case where this can help is when there are many category labels.

Faceting requires that axis limits and scales should be the same for all sub graphs.

## Univariate Charts
Univariate charts are used to charge the frequency or some summary of some event.
Univariate charts can be used as bivariate when plotting some statistical value of another variable.
This can include the mean, StdDev, median, etc.

### Qualitative Charts
Qualitative charts are used to plot frequency, or other summary value, of categorical data.

#### Bar Charts
Bar charts are typically used to charge the number of times a value took on a categorical value.
Bar charts should always have a baseline of 0 to prevent data distortion.
For nominal data, bar charts should be sorted by height.
For ordinal data, bar charts should be sorted by the ordinal.
Horizontal bar charts can be useful if there are many categories or categories have long names.

Is some cases, it is useful to track proportion of data belonging to a category in the y axis instead of the absolute counts.
The first chart uses absolute frequency while the second used relative.

#### Pie/Donut Charts
Pie charts are used for charting relative frequencies for categorical data.
Bar charts are typically more readable.
Pie charts should always represent the data as a whole.
Pie charts are only able to show a small number of categories.
Pie charts should be displayed with the first slice starting at 12 o'clock and ordered from largest to smallest in the clockwise direction.

### Quantitative Charge
Quantitative charts are used to plot frequency, or other summary value, of numeric data.

#### Histogram
Histograms group value into bins.
Bin sizes that are to small/large can make it difficult to interpret data.

#### Scatter Plots
Scatter plots show the relationship between two variables.
Scatter plots are a visual representation of the correlation coefficient.

Scatter plots can be difficult to interpret when there are too many points.
This is called overplotting.
It can be resolved with sampling or transparency.

Scatter plots can be difficult to interpret when the data charted is discrete.
This can be solved with jitter and transparency.

#### Heat Map
Heat Maps are a variations of Scatter Plots where density is mapped to a color range.
Color can be difficult to interpret so it's a good idea to add count labels to each grid square.
Heat Maps are better than Scatter Plots for discrete vs. discrete charts or when there is a lot of data to chart.
Bin size has an impart on the heat map similar to histograms.

#### Line Plots
Line Plots are used to plot the summary of one variable by the value of another.
Line Plots are good for showing relative change and emphasizes a relationship between one x value and another.
These are good for time series charts or ordinal categorical data.

#### Violin Plots
Violin plots show the range of a quantitative value as the height of the shape.
The width of the shape show how many time the variable took on that value.
The quantitative variable is grouped by the qualitative one.

#### Box Plots
Box plots are better for explanatory visualizations.

Box plots show the median value in a box. 
The top and bottom of the box represent the 3rd and 1st quantile respectively.
Whiskers represent the min and max values with a max length of (1.5 * Inter Quantile Range (IQR)).
Outliers are charted as dots past the whiskers.
Outliers are detected using the IQR as follows.

IQR = Q3 - Q1  
Outliers > Q3 + 1.5 * IQR 
Outliers < Q1 - 1.5 * IQR
